```python
cols = ['Name', 'Gender', 'Count']
file_path = os.path.join(DATA_PATH, 'yob1900.txt')

with open(file_path) as f:
    names_1900 = pd.read_csv(f, names=cols)
```

```python
file_path = os.path.join(DATA_PATH, 'yob1950.txt')

with open(file_path) as f:
    names_1950 = pd.read_csv(f, names=cols)
```

```python
names_1900_1950 = pd.merge(names_1900, names_1950, on=['Name', 'Gender'], suffixes=['_1900', '_1950'])
names_1900_1950.head(10)
```

```python
names_1900_1950.describe()
```

```python
file_path = os.path.join(DATA_PATH, 'yob2000.txt')

with open(file_path) as f:
    names_2000 = pd.read_csv(f, names=cols)
```

```python
names_all = pd.concat({
                            '1900': names_1900,
                            '1950': names_1950,
                            '2000': names_2000
                        }, names=['Year', 'Pos'])
```

```python
names_all.sort_values(by='Count', ascending=False).head(10)
```

```python
starts_with_M = names_all.Name.str.startswith('M')
names_all[starts_with_M].sort_values(by='Count', ascending=False).head(10)
```

```python
names_all.groupby('Name').sum().sort_values(by='Count', ascending=False).head(10)
```

```python
names_var_2 = names_all.groupby('Name').sum().sort_values(by='Count', ascending=False).head(10)
```

```python
names_var_2 = names_var_2.reset_index()
```
